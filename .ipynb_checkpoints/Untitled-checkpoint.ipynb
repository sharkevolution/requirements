```python
from collections import defaultdict

# https://stackoverflow.com/questions/57574409/applying-style-to-a-pandas-dataframe-row-wise
# https://github.com/epmoyer/ipy_table/issues/24

import pandas as pd

costs = {'W': {'A': 16, 'B': 16, 'C': 13, 'D': 22, 'E': 17},
         'X': {'A': 14, 'B': 14, 'C': 13, 'D': 19, 'E': 15},
         'Y': {'A': 19, 'B': 19, 'C': 20, 'D': 23, 'E': 50},
         'Z': {'A': 50, 'B': 12, 'C': 50, 'D': 15, 'E': 11}}


# Спрос - по столбцам
demand = {'A': 30, 'B': 20, 'C': 70, 'D': 30, 'E': 60}
cols = sorted(demand.keys())

# Возможности поставщиков
supply = {'W': 50, 'X': 60, 'Y': 50, 'Z': 50}

# Матрица результатов
res = dict((k, defaultdict(int)) for k in costs)

g = {}

# # Сортировка минимальной стоимости доставки в каждой строке поставщика
# for x in supply:
#     print(sorted(costs[x], key=lambda f: costs[x][f]))


t1 = pd.DataFrame(costs)
row_demand = pd.DataFrame({'demand': demand})
t1.insert(4, 'demand', row_demand, True)

t2 = pd.DataFrame.transpose(t1)
col_supply = pd.DataFrame({'supply': supply})
t2.insert(5, 'supply', col_supply, True)
t2.append(t2.sum().rename('Total'))


for b in range(5):
    t2.loc[f'drow-{b}'] = 0
    t2.loc[:, f'dcol-{b}'] = 0

def highlight_cols(s):
    color = 'lightblue'
    return 'background-color: %s' % color

def highlight_rows(s):
    color = 'lightblue'
    return 'background-color: %s' % color

print("\nHighlight specific columns:")

# t2.style.set_properties(subset = pd.IndexSlice[['demand'], :], **{'background-color' : 'yellow'})
# t2.style.set_properties(subset = pd.IndexSlice[:, ['supply']], **{'background-color' : 'yellow'})
# t2.style.applymap(highlight_cols, subset = pd.IndexSlice[:, ['supply']])

t2.style.applymap(highlight_cols, subset = pd.IndexSlice[['demand'], :])


t2.style.set_properties(subset = pd.IndexSlice[['demand'], :], **{'background-color' : 'lightblue'}) \
    .set_properties(subset = pd.IndexSlice[:, ['supply']], **{'background-color' : 'lightblue'})

```
